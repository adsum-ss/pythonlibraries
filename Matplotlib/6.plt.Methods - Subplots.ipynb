```python
# Basic Library
import numpy as np
import pandas as pd

# Libraries for Visualization
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

import warnings;
warnings.filterwarnings('ignore')
```

### Functional Method

```python
age = [25, 27, 29, 31, 33, 35, 37, 39, 41, 43, 45]
salary = [38496, 42000, 46752, 49320, 53200, 56000, 62316, 64928, 67317, 68748, 73752]
salary2 = [45372, 48876, 53850, 57287, 63016, 65998, 70003, 70000, 71496, 75370, 83640]
```

```python
plt.plot(age, salary)

plt.xlabel("age")
plt.ylabel("salary")
plt.title("Salary by Age")

plt.show()
```

### Object Oriented Method

```python
fig, ax = plt.subplots()

ax.plot(age,salary, 'r')

ax.set_xlabel("age")
ax.set_ylabel('salary')
ax.set_title("Salary by Age")

plt.show()
```

```python
# functional

plt.subplot(2,1,1)
plt.plot(age,salary, "r")

plt.subplot(2,1,2)
plt.plot(age,salary2, "b")

plt.show()
```

```python
# object-oriented

fig, ax = plt.subplots(2,1)  # or fig,ax=plt.subplots(nrows=2,ncols=1)
ax[0].plot(age,salary, 'r')
ax[1].plot(age,salary2, 'b')

plt.show()
```

```python
# use 'subplot' with functional method, and 'subplots' with object-oriented method
```

```python
# functional

plt.subplot(1,2,1)
plt.plot(age, salary, "r")

plt.subplot(1,2,2)
plt.plot(age, salary2, "b")

plt.tight_layout()
plt.show()
```

```python
# OOP

fig, ax = plt.subplots(1,2)

ax[0].plot(age,salary, 'r')
ax[1].plot(age,salary2, 'b')

plt.tight_layout(pad=0.4, w_pad=4.5, h_pad=1.0)
plt.show()
```

```python
# functional

plt.figure(figsize=(15,3))

plt.subplot(1,2,1)
plt.plot(age, salary, "r+", ls="--")

plt.subplot(1,2,2)
plt.plot(age, salary2, "bo", ls=":")

plt.show()
```

```python
# OOP

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15,3))

fig.suptitle('Horizontally stacked subplots')

ax[0].plot(age, salary, 'orange')
ax[0].set_xlabel('age')
ax[0].set_title('First')

ax[1].plot(age, salary2, 'y')
ax[1].set_title('Second')
ax[1].set_xlabel('age')

plt.tight_layout()
plt.show()
```

```python
# functional

plt.subplot(2,2,1)
plt.plot(age, salary, "r")

plt.subplot(2,2,2)
plt.plot(age, salary2, "b")

plt.subplot(2,2,3)
plt.plot(age, salary, "g")

plt.subplot(2,2,4)
plt.plot(age, salary2, "y")

plt.tight_layout();
plt.show()
```

```python
# OOP

fig, ax = plt.subplots(2,3)

ax[1][2].plot(age, salary)   # ax[1,2] 
ax[0][1].plot(age, salary)   # ax[0,1] 

plt.tight_layout()
plt.show()
```

```python
age = [25, 27, 29, 31, 33, 35, 37, 39, 41, 43, 45]
salary = [38496, 42000, 46752, 49320, 53200, 56000, 62316, 64928, 67317, 68748, 73752]

fig, axes = plt.subplots(nrows = 3, ncols = 3)
color = ['r', 'b', 'c', 'g', 'm', 'k', 'r', 'y', 'g']

a = 0
for x,y,z in axes:
    x.plot(age,salary, color[a])
    a += 1
    y.plot(age,salary, color[a])
    a += 1
    z.plot(age,salary, color[a])
    a += 1
    
plt.tight_layout()
```

```python
for i in range(1, 7):
    plt.subplot(2, 3, i)
    plt.text(0.5, 0.5, str((2, 3, i)),fontsize=18, ha='center')
    plt.title(i)

plt.suptitle("with for loop")

plt.tight_layout() 
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
for a in axes:
  a.plot(age,salary)

fig.tight_layout(w_pad=4.5)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
for a in axes:
  if (5%2==0):
    a.plot(age,salary)
  else:
    a.plot(salary,age)
    
fig.tight_layout(w_pad=4.5)
```

## rowspan and colspan

```python
# rowspan and colspan

plt.figure()

a1 = plt.subplot2grid((3,3),(0,0),colspan = 2)
a2 = plt.subplot2grid((3,3),(0,2), rowspan = 3)
a3 = plt.subplot2grid((3,3),(1,0),rowspan = 2, colspan = 2)

x = np.arange(1,10)

a1.plot(x, np.exp(x))
a1.set_title('exp')

a2.plot(x, x*x)
a2.set_title('square')

a3.plot(x, np.log(x))
a3.set_title('log')

plt.tight_layout()
plt.show()
```

## Manual spacing on subplots()
Use .subplots_adjust to adjust spacing manually.

```python
fig, ax = plt.subplots(2,2, figsize=(8,5), sharey=True, sharex=True)

ax[0,0].plot(age,salary, "r")
ax[0,0].set_title("salary by age")

ax[0,1].plot(age, salary2)
ax[0,1].set_title("salary2 by age")

ax[1,0].plot(age,salary, "g")
ax[1,0].set_title("salary by age")

ax[1,1].plot(age, salary2, "y")
ax[1,1].set_title("salary2 by age")

fig.subplots_adjust(left=None,
    bottom=None,
    right=None,
    top=None,
    wspace=0.9,
    hspace=0.5)

plt.show()
```
