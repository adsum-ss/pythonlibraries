```python
import pandas as pd
import numpy as np
```

```python
countries = {
    'country': ['Turkey', 'Cameroon', 'Albania', 'Brazil', 'Mauritius', 'Canada', 'Japan', 'Fiji'], 
    'capital': ['Ankara', 'Yaounde', 'Tirana', 'Brasilia', np.nan, 'Ottawa', 'Tokyo', 'Suva'], 
    'continent': ['Europe/Asia', 'AFRICA', 'EUROPE', 'South America',  'Africa', 'North America', 'ASIA', 'Australia/Oceania'],
    'population': [84339067, 26545863, 2877797, 212559417, np.nan, 37742154, 126476461, 896444],
    'land area': [769630, 472710, 27400, 8358140, np.nan, 9093510, 364555, 18270],
    'density': [109, 56, 105, 25, np.nan, 4, 346, 49],
    'ocean_coast': ['False', 'True', 'False', 'True', np.nan, 'True', 'True', 'True']
}
```

```python
df = pd.DataFrame(countries)
df
```

## .sort_values()

<i> Sort by the values along either axis.

```python
df["country"].sort_values()
```

```python
df.sort_values(by="country")
```

```python
df.sort_values(by="country", ignore_index=True)
```

```python
df.sort_values(by=['ocean_coast', 'continent'], ascending=[True, False])[['ocean_coast', 'continent']] 
# 'inplace=True' to modify the original dataframe itself. The return type is None.
```

```python
df.sort_values("capital", na_position='first')
```

```python
df.sort_values(by=["continent"], key=lambda x:x.str.lower())

# The key function (str.lower()) is applied to all values in "continent" column and then sorted. 
# If containing the same values, sorted based on the row index.
```

```python
df.sort_values(by=["continent"])
```

```python
df.sort_values(by='country', key=lambda x: x.str.len())
```

```python
df.sort_values(by='density')
```

```python
df[sorted(df)]
# # sorted dataframe by column names
```

```python
df[sorted(df,key=len)]
# sorted dataframe by column lengths
```

## sorting dataframe by row values / sort_index

```python
df.sort_values(by=0,axis=1)

# If the row to be sorted contains different data types, it will raise TypeError.
```

```python
df.sort_values(by=4,axis=1)
```

```python
# Sorting dataframe by row index

df.sort_index(ascending=False) # or df.sort_index(axis=0, ascending=False)
```

```python
# Sorting dataframe by column index

df.sort_index(axis=1)
```

```python
records = [(2, 33, 23, 45), (3, 4, 2, 4), (4, 5, 7, 19), (4, 6, 71, 2)]
df = pd.DataFrame.from_records(records, columns = ('0:00', '23:40', '12:30', '11:23'))
df
```

## .nlargest / .nsmallest

<i> Return the first n rows with the largest/smallest values in columns, in descending/ascending order. The columns that are not specified are returned as well, but not used for ordering.
    
<i> This method is equivalent to df.sort_values(columns, ascending=False/True).head(n), but more performant.

```python
df = pd.read_csv("Data/A.Efes 20-21 Euroleague-Points.csv")
df
```

```python
df['Age'].nlargest(8)
```

```python
df.nlargest(5, "Age", keep="all")
```

```python
df.nlargest(5, "Age", keep="last")
```

```python
df.nlargest(5, "Age", keep="first")
```

```python
df.nlargest(5, ["Age", "Height"])
```

```python
df.nsmallest(5, ["Age"])
```
