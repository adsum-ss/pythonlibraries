## Plotting Styles

There are many other plotting styles available other than the default Line plot, all of which can be accessed by passing kind keyword to plot(). The full list of available plots are as follows:
<li>bar for vertical bar plots</li>
<li>barh for horizontal bar plots</li>
<li>hist for histogram</li>
<li>box for boxplot</li>
<li>kde or density for density plots</li>
<li>area for area plots</li>
<li>pie for pie plots</li>
<li>scatter for scatter plots</li>
<li>hexbin for hexbin plot</li>

### Figure Size

```python
# functional method

plt.figure(figsize=(13,3))

xpoints = np.array([1,2,6,8,])
ypoints = np.array([3,8,1,10])

plt.plot(xpoints, ypoints);
```

```python
# object-oriented method

fig, ax = plt.subplots(figsize=(13,3))

xpoints = np.array([1,2,6,8,])
ypoints = np.array([3,8,1,10])

ax.plot(xpoints, ypoints);
```

## xlim and ylim

```python
fig, ax = plt.subplots()

m = np.arange(1,10)
n = m ** 2

a = ax.bar(m,n)

ax.bar_label(a, padding=3)

plt.title('Numbers')
plt.show()
```

```python
fig, ax = plt.subplots()

m = np.arange(1,10)
n = m ** 2

a = ax.bar(m,n)

ax.bar_label(a, padding=3)

plt.xlim(0, 12)   # axes.set_xlim for object oriented method
plt.ylim(0,100)   # axes.set_ylim for object oriented method

plt.title('Numbers')
plt.show()
```

## tight_layout

```python
fig, axes = plt.subplots(nrows=3,ncols=2)
```

```python
fig, axes = plt.subplots(nrows=3,ncols=2)

plt.tight_layout()
```

```python
def example_plot(ax, fontsize=12):
    ax.plot([1, 2])

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2)

example_plot(ax1)
example_plot(ax2)
example_plot(ax3)
example_plot(ax4)

plt.tight_layout(pad=0.4, w_pad=4.5, h_pad=2.5)
```

## colorbar

```python
plt.close('all')
arr = np.arange(100).reshape((10, 10))
fig = plt.figure(figsize=(4, 4))
im = plt.imshow(arr, interpolation="none")

plt.colorbar(im)

plt.tight_layout()
```
