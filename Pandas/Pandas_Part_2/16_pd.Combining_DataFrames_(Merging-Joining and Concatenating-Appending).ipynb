```python
import pandas as pd
```

```python
class display(object):
    """Display HTML representation of multiple objects"""
    template = """<div style="float: left; padding: 10px;">
    <p style='font-family:"Courier New", Courier, monospace'>{0}</p>{1}
    </div>"""
    def __init__(self, *args):
        self.args = args
        
    def _repr_html_(self):
        return '\n'.join(self.template.format(a, eval(a)._repr_html_())
                         for a in self.args)
    
    def __repr__(self):
        return '\n\n'.join(a + '\n' + repr(eval(a))
                           for a in self.args)
```

## pd.merge()

### first dataframe example

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                     'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']})
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
# inner = pd.merge(left,right) # inner join
# how{‘left’, ‘right’, ‘outer’, ‘inner’, ‘cross’}, default ‘inner’

display("left", "right", "pd.merge(left,right)", "pd.merge(left,right, on='key')")
```

```python
# outer = pd.merge(left, right, on="key", how="outer")

display("left", "right", "pd.merge(left,right, on='key', how='outer')")
```

```python
# left = pd.merge(left, right, on="key", how="left")

display("left", "right", "pd.merge(left,right, on='key', how='left')")
```

```python
# right = pd.merge(left, right, on="key", how="right")

display("left", "right", "pd.merge(left,right, on='key', how='right')")
```

### second dataframe example

```python
df1 = pd.DataFrame({"ID": [1,2,3,5,9],
                    "Col1": [1,2,3,4,5],
                    "Col2": [6,7,8,9,10],
                    "Col3": [11,12,13,14,15],
                    "Col4": ['apple','orange','banana','strawberry','raspberry']
                   })

df2 = pd.DataFrame({"ID": [1,1,3,5],
                    "ColA": [8,9,10,11],
                    "ColB": [12,13,15,17],
                    "Col4": ['apple','orange','banana','kiwi'],
                   })
```

```python
# pd.merge(df1, df2, suffixes=['_l', '_r'], left_on='Col2', right_on='ColA')

display("df1", "df2", "pd.merge(df1, df2, suffixes=['_l', '_r'], left_on='Col2', right_on='ColA')")
```

```python
# pd.merge(df1, df2, suffixes=['_l', '_r'], left_index=True, right_index=True)
# joining on indices

display("df1", "df2", "pd.merge(df1, df2, suffixes=['_l', '_r'], left_index=True, right_index=True)")
```

### pd.merge -- join

<i> <b>inner join:</b> Joins all shared rows on the joining/key columns. Rows are lost that don't have a match in the other DataFrame's key column.
    <br><b>outer join:</b> Joins all rows from both DataFrames. No data will be lost.    
    <b>left join:</b> Joins on all rows from the left DataFrame. Any rows from the right DataFrame that do not have a match in the key column of the left DataFrame are discarded.    
    <b>right join:</b> The opposite of the left join. Joins on all rows from the right DataFrame.

```python
# inner join -- pd.merge(df1,df2)

display("df1", "df2", "pd.merge(df1, df2)")
```

```python
# outer join -- pd.merge(df1, df2, on='Col4', how='outer', suffixes=['_l', '_r'])

display("df1", "df2", "pd.merge(df1, df2, on='Col4', how='outer', suffixes=['_l', '_r'])")
```

```python
# left join -- pd.merge(df1, df2, on='Col4', how='left', suffixes=['_l', '_r'])

display("df1", "df2", "pd.merge(df1, df2, on='Col4', how='left', suffixes=['_l', '_r'])")
```

```python
# right join -- pd.merge(df1, df2, on='Col4', how='right', suffixes=['_l', '_r'])

display("df1", "df2", "pd.merge(df1, df2, on='Col4', how='right', suffixes=['_l', '_r'])")
```

### Common issues/errors

```python
pd.merge(df1,df2,on='Col_1')
# key error because 'Col_1' is not in both DataFrame
```

```python
pd.merge(df1,df2,on='Col_4')
# "Col_4": [1,2,3,4]
# value error because the joining column does not have a consitent dtype
```

```python
pd.merge(df1, df2, on='ID', suffixes=['_l', '_r'], left_on='Col_2', right_on='Col_A')
# merge error because we have 'on' and 'left_on&right_on' included as argument
```

# df.join

<i> anything you can do with .join you can do with .merge
    
<i><b>.merge is more powerful than .join

### first dataframe example

```python
df3 = pd.DataFrame({'key': ['K0', 'K2', 'K3', 'K4', 'K5', 'K6'],
                    'X': ['X0', 'X2', 'X3', 'X4', 'X5', 'X6']})
df4 = pd.DataFrame({'key': ['K0', 'K2', 'K3'],
                    'Y': ['Y0', 'Y2', 'Y3']})
```

```python
# df4.join(other)  gives an error
```

```python
# inner = df3.join(df4, lsuffix='_df', rsuffix='_other', how='inner')
# how{‘left’, ‘right’, ‘outer’, ‘inner’}, default ‘left’

display("df3", "df4", "df3.join(df4, lsuffix='_df3', rsuffix='_df4', how='inner')")
```

```python
# outer = df3.join(df4, lsuffix='_df', rsuffix='_other', how='outer')

display("df3", "df4", "df3.join(df4, lsuffix='_df3', rsuffix='_df4', how='outer')")
```

```python
# right = df3.join(df4, lsuffix='_df', rsuffix='_other', how='right')

display("df3", "df4", "df3.join(df4, lsuffix='_df3', rsuffix='_df4', how='right')")
```

```python
# left = df3.join(df4, lsuffix='_df', rsuffix='_other')

display("df3", "df4", "df3.join(df4, lsuffix='_df3', rsuffix='_df4')")
```

```python
df4.set_index("key")
# df3.set_index("key")
```

```python
df3.join(df4.set_index("key"), on="key")  
# In such cases, it makes more sense to use .merge()
```

# pd.concat()

<i> good for side-by-side and stacking DataFrames on top of each other.
    
<i> with concatenation, your datasets are just stitched together along an axis --either the row axis or column axis.

```python
# pd.concat([df1, df2])
# default axis is 0 (this stacks the DataFrames)

display("df1", "df2", "pd.concat([df1, df2])")
```

```python
# pd.concat([df1, df2], ignore_index=True)
# reset the index

display("df1", "df2", "pd.concat([df1, df2], ignore_index=True)")
```

```python
# pd.concat([df1, df2], axis=1)
# can side-by-side by specifying axis=1

display("df1", "df2", "pd.concat([df1, df2], axis=1)")
```

```python
# pd.concat([df1, df2], axis=1, join='inner')

# default join is outer, but you can specify inner join
# there is no option for left or right joins

display("df1", "df2", "pd.concat([df1, df2], axis=1, join='inner')")
```

```python
# pd.concat([df1, df2], axis=0, join='inner')

display("df1", "df2", "pd.concat([df1, df2], axis=0, join='inner')")
```
