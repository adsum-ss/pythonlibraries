## boxenplot (letter value plot)

Draw an enhanced box plot for larger datasets.

This style of plot was originally named a “letter value” plot because it shows a large number of quantiles that are defined as “letter values”. It is similar to a box plot in plotting a nonparametric representation of a distribution in which all features correspond to actual observations. By plotting more quantiles, it provides more information about the shape of the distribution, particularly in the tails.

```python
# Basic Library
import numpy as np
import pandas as pd

# Libraries for Visualization
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

import scipy
import warnings;
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("../Data/StudentsPerformance.csv")
df.head()
```

```python
df.shape
```

```python
sns.boxenplot(x=df['math score'])
```

```python
sns.boxenplot(x='math score', y='race/ethnicity', data=df);
```

```python
sns.boxenplot(x='race/ethnicity',y='math score',data=df);
```

```python
plt.figure(figsize=(12,6))
sns.boxenplot(x='race/ethnicity',y='math score',data=df,hue='gender',palette="Set3",linewidth=2.5);
```

```python
## Order

plt.figure(figsize=(12,6))
sns.boxenplot(x='race/ethnicity',y='math score',data=df,
              order=["group A", "group B", "group C", "group D", "group E"]);
```

### Draw a boxen plot for each numeric variable in a DataFrame:

```python
iris = sns.load_dataset("iris")
ax = sns.boxenplot(data=iris, orient="h", palette="Set2")
```

### Use stripplot() to show the datapoints on top of the boxes:

```python
plt.figure(figsize=(10,5))

ax = sns.boxenplot(x="race/ethnicity", y="math score", data=df, showfliers=False)
ax = sns.stripplot(x="race/ethnicity", y="math score", data=df, size=4, color=".26")
```

### Use catplot() to combine boxenplot() and a FacetGrid. This allows grouping within additional categorical variables. Using catplot() is safer than using FacetGrid directly, as it ensures synchronization of variable order across facets:

```python
g = sns.catplot(x="test preparation course", y="math score",
                hue="gender", col="race/ethnicity",
                data=df, kind="boxen",
                height=4, aspect=.7);
```
