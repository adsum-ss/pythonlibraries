## Scatter Plot

Scatter plots show the relationship between two variables(continuous numerical) in the form of dots on the plot. 
In simple terms, the values along a horizontal axis are plotted against a vertical axis.

<i>matplotlib.pyplot.scatter(x, y, s=None, c=None, marker=None, cmap=None, norm=None, vmin=None, vmax=None, alpha=None, linewidths=None, *, edgecolors=None, plotnonfinite=False, data=None, **kwargs)[source]

```python
# Basic Library
import numpy as np
import pandas as pd

# Libraries for Visualization
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

import warnings;
warnings.filterwarnings('ignore')
```

```python
# s ===> The marker size in points
# c ===> The marker colors
# alpha ===> The alpha blending value, between 0 (transparent) and 1 (opaque)
# marker ===> The marker style
```

```python
y = np.random.randint(10,100,100)
x = np.random.randint(10,100,100)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y, s=20, color='green', marker='d', alpha=0.25)
plt.show()
```

```python
area = [i for i in range(len(x))]
colors = np.random.rand(100)
```

```python
plt.scatter(x, y, s=area, c=colors, alpha=1)
plt.show()
```

```python
np.random.seed(19680801)

a = np.random.rand(50)
b = np.random.rand(50)
z = np.sqrt(a**2 + b**2)

fig, axs = plt.subplots(2, 3, sharex=True, sharey=True, figsize=(12,4))

axs[0, 0].scatter(a, b, s=10, c=z, marker=">")
axs[0, 0].set_title("marker='>'")

axs[0, 1].scatter(a, b, s=30, c=z, marker=r'$\alpha$')
axs[0, 1].set_title(r"marker=r'\$\alpha\$'")

verts = [[-1, -1], [1, -1], [1, 1], [-1, -1]]
axs[0, 2].scatter(a, b, s=50, c=z, marker=verts)
axs[0, 2].set_title("marker=verts")

axs[1, 0].scatter(a, b, s=70, c=z, marker=(5, 0))
axs[1, 0].set_title("marker=(5, 0)")

axs[1, 1].scatter(a, b, s=80, c=z, marker=(5, 1))
axs[1, 1].set_title("marker=(5, 1)")

axs[1, 2].scatter(a, b, s=80, c="Purple", marker=(5, 2))
axs[1, 2].set_title("marker=(5, 2)")

plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv("../Data/Final_Four 2020-2021-2.csv")
df
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10,5))

axs[0, 0].scatter(df['Assists'], df['Points'])
axs[0, 0].set_xlabel("Assists")
axs[0, 0].set_ylabel("Points")

axs[0, 1].scatter(df['Rebounds'], df['Points'])
axs[0, 1].set_xlabel("Rebounds")
axs[0, 0].set_ylabel("Points")

axs[1, 0].scatter(df['Turnovers'], df['Points'])
axs[1, 0].set_xlabel("Turnovers")
axs[0, 0].set_ylabel("Points")

axs[1, 1].scatter(df['Steals'], df['Points'])
axs[1, 1].set_xlabel("Steals")
axs[0, 0].set_ylabel("Points")

plt.tight_layout()
plt.show()
```
