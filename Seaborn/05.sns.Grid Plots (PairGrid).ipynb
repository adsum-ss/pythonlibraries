## PairGrid

Subplot grid for plotting pairwise relationships in a dataset.

This object maps each variable in a dataset onto a column and row in a grid of multiple axes. Different axes-level plotting functions can be used to draw bivariate plots in the upper and lower triangles, and the the marginal distribution of each variable can be shown on the diagonal.

Several different common plots can be generated in a single line using pairplot(). Use PairGrid when you need more flexibility.

```python
# Basic Library
import numpy as np
import pandas as pd

# Libraries for Visualization
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

import scipy

import warnings;
warnings.filterwarnings('ignore')
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
g = sns.PairGrid(tips);
```

```python
type(g)
```

```python
g = sns.PairGrid(tips)
g.map(sns.scatterplot);
```

```python
sns.set_style('darkgrid')
g = sns.PairGrid(tips)
g.map(sns.histplot);
```

```python
# Matplotlib Functions

g = sns.PairGrid(tips)
g.map(plt.scatter);
```

```python
g = sns.PairGrid(tips)
g.map_offdiag(sns.scatterplot)
g.map_diag(sns.histplot);
```

```python
g = sns.PairGrid(tips)

g.map_upper(sns.scatterplot)
g.map_diag(sns.histplot);
```

```python
g = sns.PairGrid(tips)

g.map_upper(sns.scatterplot)
g.map_diag(sns.boxplot)
g.map_lower(sns.kdeplot, colors='red');
```

```python
g = sns.PairGrid(tips, corner=True)
g.map_offdiag(sns.scatterplot)
g.map_diag(sns.boxplot);
```

```python
g = sns.PairGrid(tips, hue='sex', diag_sharey=False)
g.map_diag(sns.kdeplot) 
g.map_offdiag(sns.scatterplot);
```

```python
g = sns.PairGrid(tips, hue='sex', diag_sharey=False, vars=['total_bill', 'tip'])
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.scatterplot);
```

```python
g = sns.PairGrid(tips,  
                 x_vars=['total_bill', 'tip', 'size'],
                 y_vars=['total_bill']
                )
g.map(sns.scatterplot);
```

```python
g = sns.PairGrid(tips, hue="sex", palette="viridis")

g.map_upper(sns.scatterplot, linewidths=1, edgecolor="w", s=40)
g.map_diag(sns.distplot)
g.map_lower(sns.kdeplot)
g.add_legend();
```

```python
g = sns.PairGrid(tips, hue='sex', diag_sharey=False)
g.map_diag(sns.kdeplot, fill=True, lw=3)
g.map_offdiag(sns.scatterplot, s=80, alpha=.5, ec='black');
```

```python
g = sns.PairGrid(tips, hue='sex', diag_sharey=False, palette='viridis', height=1.5, aspect=2)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.scatterplot);
```

```python
g = sns.PairGrid(tips, vars=['total_bill', 'tip', 'size'], diag_sharey=False)

g.map_lower(sns.scatterplot)
g.map_upper(sns.kdeplot, n_levels=5)

for var in g.axes:
    for ax in var:
        ax.set_xlabel('')
        ax.set_ylabel('')
        ax.set_xticks([])
        ax.set_yticks([])
```

```python
g = sns.PairGrid(tips, 
                 hue='sex', 
                 palette='magma', 
                 vars=['total_bill', 'tip', 'size'], 
                 height=1.5, 
                 aspect=1.65, 
                 diag_sharey=False
                )
g.map_lower(sns.scatterplot, s=50, alpha=0.7, ec='gray')
g.map_diag(sns.histplot,hue=None, color='navy')
g.map_upper(sns.kdeplot, n_levels=4,hue=None, color='mediumvioletred')
for var in g.axes:
    for ax in var:
        ax.set_xlabel('')
        ax.set_ylabel('')
        ax.set_xticks([])
        ax.set_yticks([])
#g.savefig('pairgrid_example.png', dpi=200);
```
