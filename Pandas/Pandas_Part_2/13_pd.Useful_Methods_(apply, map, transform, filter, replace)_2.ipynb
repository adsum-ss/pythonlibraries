```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib inline
```

```python
df = pd.DataFrame({'groups': ['A', 'B', 'C', 'A', 'B', 'C','A', 'B', 'C'],
                   'var1': [10,20,30,40,50,60,70,80,90],
                   'var2': [100,200,300,400,500,600,700,800,900]})
df
```

```python
df[["var1"]].transform(lambda x : ( x - df["var1"].min() )/
                                        (df["var1"].max()-df["var1"].min())
                       )
# makes minmaxscaler

# df[["var1"]].transform(lambda x : (x - x.min()) / (x.max() - x.min()))
```

```python
df[["var1"]].apply(lambda x : ( x - df["var1"].min() )/
                                    (df["var1"].max()-df["var1"].min())
                   )
# makes minmaxscaler

# df[["var1"]].apply(lambda x : (x - x.min()) / (x.max() - x.min()))
```

```python
df[["var1","var2"]].transform(lambda x: (x-x.min())/(x.max()-x.min()))
# makes minmaxscaler
```

```python
df[["var1","var2"]].apply(lambda x: (x-x.min())/(x.max()-x.min()))
# makes minmaxscaler
```
