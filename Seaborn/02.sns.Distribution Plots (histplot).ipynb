## Histogram

A histogram is a classic visualization tool that represents the distribution of one or more variables by counting the number of observations that fall within disrete bins.

This function can normalize the statistic computed within each bin to estimate frequency, density or probability mass, and it can add a smooth curve obtained using a kernel density estimate

<i>seaborn.histplot(data=None,x=None, y=None, hue=None, bins='auto', kde=False)

```python
# Basic Library
import numpy as np
import pandas as pd

# Libraries for Visualization
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

import warnings;
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Final_Four 2020-2021.csv')
```

```python
sns.distplot(df.Points);  # distplot has now been deprecated. Switch to histplot for histogram plots.
```

```python
sns.histplot(df.Height);
```

```python
plt.figure(figsize=(7,4))

sns.histplot(df.Height, alpha=1)

sns.despine(left=True)

plt.yticks([])
plt.xlabel('')
plt.ylabel('')
plt.tight_layout();
```

```python
sns.set_style('darkgrid')

sns.histplot(x='Height', data=df);
```

```python
sns.histplot(y='Height', data=df);
```

```python
# kde

sns.histplot(x='Height', data=df, kde=True);
```

```python
# bins

sns.histplot(x='Height', data=df, bins=20);
```

```python
sns.set(style='darkgrid', rc={"grid.color": ".6", "grid.linestyle": ":"})
sns.histplot(data = df, x = 'Height', bins=20);
```

```python
sns.histplot(x='Height', data=df, bins=[180, 190, 205, 215, 225]);
```

```python
# binwidth and binrange

sns.histplot(x='Height', data=df, binwidth=5, binrange=(175, 230));
```

```python
# STATISTICS, stat

sns.histplot(x='Height', data=df, stat='count');
```

```python
sns.histplot(x='Height', data=df, stat='density');
```

```python
sns.histplot(x='Height', data=df, stat='probability');
```

```python
sns.histplot(x='Height', data=df, stat='probability', fill=False, element='step', cumulative=True);
```

```python
# Categorical Variables (hue)

sns.histplot(x='Height', data=df, hue='Position');
```

```python
sns.histplot(x='Height', data=df, hue='Position', element='step');
```

```python
sns.histplot(x='Height', data=df, hue='Position', element='poly');
```

```python
# multiple

sns.histplot(x='Height', data=df, hue='Position', multiple='stack');
```

```python
sns.histplot(x='Height', data=df, hue='Position', multiple='fill');
```

```python
sns.histplot(x='Height', data=df, hue='Position', multiple='dodge');
```

```python
sns.histplot(x='Height', data=df, hue='Position', multiple='dodge', shrink=0.3);
```

```python
sns.histplot(x='Height', y='Rebounds',data=df, cbar=True);
```

```python
# color, palette

sns.histplot(x='Height', data=df, color='indigo');
```

```python
plt.figure(figsize=(8,4))
sns.histplot(x='Club', hue='Position', multiple='dodge', data=df, shrink=0.8, palette='bone');
```

```python
# fill

sns.histplot(x='Height', data=df, fill=False);
```

## EXAMPLES

```python
cars = sns.load_dataset('mpg').dropna()
cars.head()
```

```python
sns.set_style('whitegrid')

hp = cars.horsepower

plt.figure(figsize=(8,5))
sns.distplot(hp, kde_kws={"lw": 5}, hist_kws={'alpha': 0.25})

sns.despine(left=True)
plt.xlim(0, 280)
plt.xlabel('')
plt.yticks([])
plt.tight_layout();
```
