```python
# Basic Library
import numpy as np
import pandas as pd

# Libraries for Visualization
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

import warnings;
warnings.filterwarnings('ignore')
```

```python
age = [25, 27, 29, 31, 33, 35, 37, 39, 41, 43, 45]
salary = [38496, 42000, 46752, 49320, 53200, 56000, 62316, 64928, 67317, 68748, 73752]
salary2 = [45372, 48876, 53850, 57287, 63016, 65998, 70003, 70000, 71496, 75370, 83640]
```

```python
fig = plt.figure()                        # ax = fig.add_axes([left, bottom, widht, height])
                                          # left, bottom, width, height (range 0 to 1)
ax1 = fig.add_axes([0.0, 0.0, 1, 1])
ax2 = fig.add_axes([1.1, 0.3, 0.4, 0.4])

ax1.plot(age,salary, "r")
ax1.set_xlabel("Age")
ax1.set_ylabel("Salary")
ax1.set_title("Salary by Age")

ax2.plot(age,salary2, "b");
ax2.set_xlabel("Age")
ax2.set_ylabel("Salary2")
ax2.set_title("Salary2 by Age");

plt.show()
```

```python
fig = plt.figure()                        # ax = fig.add_axes([left, bottom, widht, height])
                                          # left, bottom, width, height (range 0 to 1)
ax1 = fig.add_axes([0.0, 0.0, 2, 1])
ax2 = fig.add_axes([1.1, 0.3, 0.4, 0.4])

ax1.plot(age,salary, "r")
ax1.set_xlabel("Age")
ax1.set_ylabel("Salary")
ax1.set_title("Salary by Age")

ax2.plot(age,salary2, "b");
ax2.set_xlabel("Age")
ax2.set_ylabel("Salary2")
ax2.set_title("Salary2 by Age");

plt.show()
```

```python
fig = plt.figure()

a = np.arange(0,10)
b = a**2

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(a,b)
axes2.plot(b,a)

axes1.set_title('LARGER')
axes2.set_title('SMALLER');
```

```python
fig = plt.figure()

x = np.arange(0,11)
y = x * 2

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,label='xvsy')
axes.plot(y,x,label='yvsx')

axes.legend(loc=(0.5,0.5));
```

```python
fig = plt.figure(figsize=(3,2), dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```
