```python
# Basic Library
import numpy as np
import pandas as pd

# Libraries for Visualization
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

import warnings;
warnings.filterwarnings('ignore')
```

```python
age = [25, 27, 29, 31, 33, 35, 37, 39, 41, 43, 45]
salary = [38496, 42000, 46752, 49320, 53200, 56000, 62316, 64928, 67317, 68748, 73752]
salary2 = [45372, 48876, 53850, 57287, 63016, 65998, 70003, 70000, 71496, 75370, 83640]
```

## Multiple-Line in the same graph

### Plot Range and Adding Extra Lines

```python
plt.plot(age, salary)
plt.plot(age, salary2)

plt.xlabel("age")
plt.ylabel("salary")

plt.title("Salary by Age")
plt.show()
```

```python
x = np.arange(0,11)
y = x.copy()
z = x*2
t = np.log(x)
```

```python
fig, ax = plt.subplots()

ax.plot(x, y)
ax.plot(x, z)
ax.plot(x, t)

ax.set_xlabel("x")
ax.set_ylabel("y - z - t")
ax.set_title("title");

plt.show()
```

```python
# Plot Range with set_ylim & set_xlim

x = np.arange(1,10)

fig, ax = plt.subplots(nrows=1,ncols=2, figsize=(10,4))

ax[0].plot(x,x**2,x,x**3,x,x**4, 'r')

ax[1].plot(x,x**2,x,x**3,x,x**4, 'r')
ax[1].set_xlim([1,3])
ax[1].set_ylim([1,50])

plt.show()
```

```python
# horizontal and vertical lines

x  =np.arange(1,10)

fig, ax = plt.subplots(nrows=1,ncols=2, figsize=(10,4))

ax[0].plot(x,x**2,x,x**3,x,x**4, 'r')

ax[1].plot(x,x**2,x,x**3,x,x**4, 'r')
ax[1].set_xlim([1,3])
ax[1].set_ylim([1,50])

ax[0].axvline(x=5,ls="--")
ax[1].axhline(y=15,ls="--")

plt.show()
```

```python
x = np.arange(1,10)

fig, ax = plt.subplots(nrows=1,ncols=2, figsize=(10,4))

ax[0].plot(x,x**2,x,x**3,x,x**4, 'r')
ax[0].axvline(x=2,ls="--")
ax[0].axhline(y=20,ls="--")

ax[1].plot(x,x**2,x,x**3,x,x**4, 'r')
ax[1].set_xlim([1,3])
ax[1].set_ylim([1,50])

ax[1].axvline(x=2,ls="--")
ax[1].axhline(y=20,ls="--")

plt.show()
```

```python
x = np.linspace(0,5,100)
y = np.sin(4*x)*np.exp(-x)
z = np.exp(-x)*np.cos(4*x)

plt.figure()
plt.plot(x, y, 'r--', label='$e^{-x} sin(4x)$', linewidth=4.0)
plt.plot(x, z, label='$e^{-x} cos(4x)$', linewidth=2.0)

plt.xlabel('x')
plt.ylabel('value')

plt.legend()

plt.title('My plot');
# plt.savefig('myFigure.png')
```
