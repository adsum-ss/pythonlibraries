## jointplot

Draw a plot of two variables with bivariate and univariate graphs.

This function provides a convenient interface to the JointGrid class, with several canned plot kinds. This is intended to be a fairly lightweight wrapper; if you need more flexibility, you should use JointGrid directly.

```python
# Basic Library
import numpy as np
import pandas as pd

# Libraries for Visualization
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

import scipy

import warnings;
warnings.filterwarnings('ignore')
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, color='#643db8');
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, hue='sex', color='#643db8');
```

```python
# kind : { "scatter" (default) | "kde" | "hist" | "hex" | "reg" | "resid" }
```

### kind = 'hex'

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex');
```

### kind = 'reg'

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind="reg");
```

### kind = 'kde'

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind="kde");
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, hue='sex', kind="kde");
```

### kind = 'hist'

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind="hist");
```

### kind = 'resid'

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind="resid");
```

### styling

```python
sns.set_style('darkgrid')
sns.jointplot(x='total_bill', y='tip', data=tips, color='green');
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, hue='sex', palette='autumn');
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips,
              height=8,   #Default: 6
              space=0.1,  #Default: 0.2
              ratio=2     #Default: 5
             );
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, joint_kws={'marker': 5});
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg',
              joint_kws={'ci': None, 'order': 2});
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips,
              marginal_kws={'color': 'xkcd:golden'});
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde',
              marginal_kws={'lw': 4, 'shade': True});
```

```python
sns.jointplot(data=tips, x="total_bill", y="tip",
              marker="+", s=100, marginal_kws=dict(bins=25, fill=False),
             );
```

```python
g = sns.jointplot(data=tips, x="total_bill", y="tip")
g.plot_joint(sns.kdeplot, color="r", zorder=0, levels=6)
g.plot_marginals(sns.rugplot, color="r", height=-.15, clip_on=False);
```

## EXAMPLES

```python
geyser = sns.load_dataset('geyser')
geyser.head()
```

```python
plt.rc('xtick', labelsize=12)
plt.rc('ytick', labelsize=12)
plt.rc('axes', labelsize=14)
```

```python
j = sns.jointplot(x='waiting', y='duration', 
                  data=geyser.iloc[:, :-1]+np.random.normal(0, 0.15, size=geyser.iloc[:, :-1].shape), #adding "jitter"
                  color='white')
j.plot_marginals(sns.rugplot, height=.5, color='#643db8')
j.ax_joint.set_xlabel('')
j.ax_joint.set_ylabel('');
```

```python
j = sns.jointplot(x='waiting', y='duration', 
                  data=geyser.iloc[:, :-1]+np.random.normal(0, 0.15, size=geyser.iloc[:, :-1].shape), #adding "jitter"
                  color= 'green')
j.plot_marginals(sns.rugplot, height=-.15, color='black', clip_on=False)
j.plot_joint(sns.kdeplot, color='black', levels=5)
j.ax_joint.set_xlabel('')
j.ax_joint.set_ylabel('')
j.ax_joint.set_ylim(0.5, 6)
j.ax_joint.set_xlim(30, 110);
```
