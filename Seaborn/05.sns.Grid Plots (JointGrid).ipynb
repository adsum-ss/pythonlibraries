## JointGrid

Grid for drawing a bivariate plot with marginal univariate plots.

Many plots can be drawn by using the figure-level interface jointplot(). Use this class directly when you need more flexibility

```python
# Basic Library
import numpy as np
import pandas as pd

# Libraries for Visualization
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

import scipy

import warnings;
warnings.filterwarnings('ignore')
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.set_style('darkgrid')
sns.JointGrid(data=tips, x="total_bill", y="tip");
```

```python
g = sns.JointGrid();
```

```python
g = sns.JointGrid(data=tips, x="total_bill", y="tip")
g.plot(sns.scatterplot, sns.histplot);
```

```python
g = sns.JointGrid(data=tips, x="total_bill", y="tip")
g.plot(sns.regplot, sns.boxplot);
```

```python
g = sns.JointGrid(data=tips, x="total_bill", y="tip", hue='sex')
g.plot(sns.scatterplot, sns.histplot)
g.refline(x=20, y=3);
```

```python
g = sns.JointGrid(data=tips, x="total_bill", y="tip")
g.plot(sns.scatterplot, sns.histplot, alpha=.5, edgecolor="black", linewidth=.5);
```

```python
g = sns.JointGrid(data=tips, x="total_bill", y="tip")
g.plot_joint(sns.scatterplot, s=100, alpha=.5)
g.plot_marginals(sns.histplot, kde=True);
```

```python
g = sns.JointGrid(data=tips, x="total_bill", y="tip")
g.plot_joint(sns.regplot, line_kws={'color': 'black'})
g.plot_marginals(sns.kdeplot, fill=True, color='purple');
```

```python
g = sns.JointGrid()
x, y = tips["total_bill"], tips["tip"]

sns.scatterplot(x=x, y=y, ec="b", fc="none", s=100, linewidth=1.5, ax=g.ax_joint)
sns.histplot(x=x, fill=False, linewidth=2, ax=g.ax_marg_x)
sns.kdeplot(y=y, linewidth=2, ax=g.ax_marg_y);
```

```python
g = sns.JointGrid()
x, y = tips["total_bill"], tips["tip"]

sns.regplot(x=x, y=y, ax=g.ax_joint)
sns.kdeplot(x=x, ax=g.ax_marg_x, hue=tips.sex)  
sns.histplot(y=y, ax=g.ax_marg_y);

# #g.savefig('Joint_Grid_Example.png');
```

```python
sns.set_style('white')

g = sns.JointGrid()
x,y = tips.total_bill, tips.tip
sns.regplot(x=x, y=y, ax=g.ax_joint, ci=99, 
            scatter_kws={'color': 'xkcd:lavender', 'alpha': 0.7, 'ec': 'gray', 's': 50}, 
            line_kws={'color': 'black', 'lw': 3}
           )
sns.kdeplot(x=x, ax=g.ax_marg_x, fill=True, color='xkcd:indigo', lw=2 )
sns.histplot(y=y, bins=15, ax=g.ax_marg_y, lw=2, color='xkcd:indigo', alpha=0.5, ec='xkcd:indigo')
g.set_axis_labels(xlabel='', ylabel='')

ax = g.ax_joint
ax.set_xticklabels([])
ax.set_yticklabels([]);
#g.savefig('25jointgrid_image_thumb.png', dpi=200);
```
