```python
sns.scatterplot(x='Minutes', y='Points', hue='Position', data=df, 
                palette=['purple', '#55CCCC', 'orange'], hue_order=['Guard', 'Center', 'Forward']);
```

```python
# style

sns.scatterplot(x='Minutes', y='Points', style='Position', data=df, s=75);
```

```python
# size

sns.scatterplot(x='Minutes', y='Points', size='Position', data=df, sizes=[150, 100, 50]);
```

```python
# alpha (transparency, between 0-1)

sns.scatterplot(x='Minutes', y='Points', data=df, s=100, alpha=0.2);
```

```python
# marker

sns.scatterplot(x='Minutes', y='Points', data=df, s=150, marker='*', edgecolor='black');
```

## EXAMPLES

```python
pts_x = [1, 2, 1.5, 4, 3, 2.5, 2.1, 1.5, 3.5, 1, 2.6, 3.4, 1.3, 0.5]
pts_y = [1, 3, 2.5, 3, 3.5, 2, 2.5, 1.8, 4, 1.3, 3.2, 3, 2, 1]

cat = [0, 0, 1, 1, 0, 0, 1, 0, 1, 1, 0, 0, 1, 1]
```

```python
for x,y in zip(pts_x, pts_y):
    print((x, y))
```

```python
plt.figure(figsize=(8, 5))

sns.scatterplot(pts_x, pts_y, s=200, hue=cat, alpha=0.7, edgecolor='black', palette={0: 'xkcd:crimson', 1: '#3d9f98'})
sns.despine()

plt.xlim(0, 4.2)
plt.ylim(0, 4.2)

plt.yticks([0, 1, 2, 3, 4])
plt.xticks([1, 2, 3, 4])

plt.gca().get_legend().remove()
plt.tight_layout()
# plt.savefig('08scatter_byHue.png')
```

```python
plt.figure(figsize=(8, 5))

sns.scatterplot(pts_x, pts_y, size=cat, alpha=0.7, edgecolor='black', color='xkcd:crimson', sizes=[50, 300])
sns.despine()

plt.xlim(0, 4.2)
plt.ylim(0, 4.2)

plt.yticks([0, 1, 2, 3, 4])
plt.xticks([1, 2, 3, 4])

plt.gca().get_legend().remove()
plt.tight_layout()
# plt.savefig('08scatter_bySize.png')
```

```python
plt.figure(figsize=(8, 5))

sns.scatterplot(pts_x, pts_y, size=cat, hue=cat, style=cat,
                alpha=0.7, edgecolor='black',  
                sizes=[100, 300],
                palette={0: 'xkcd:crimson', 1: '#3d9f98'}
               )
sns.despine()

plt.xlim(0, 4.2)
plt.ylim(0, 4.2)

plt.yticks([0, 1, 2, 3, 4])
plt.xticks([1, 2, 3, 4])

plt.gca().get_legend().remove()
plt.tight_layout()
# plt.savefig('08scatter_byAll.png')
```
