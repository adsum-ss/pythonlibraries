## <b> contains ()

<i> Test if pattern or regex is contained within a string of a Series or Index.

<i> Return boolean Series or Index based on whether a given pattern or regex is contained within a string of a Series or Index.

```python
df
```

```python
df[df["Position"] == "Forward"]['Country'].value_counts()
```

```python
filt = df['Position'] == 'Forward'
df.loc[filt]['Country'].str.contains('Turkey').sum()
```

```python
filt = df['Country'].str.contains('s', na=False)
filt
# when we have null values in our dataset; if we don't use "na", we get error; 
# with "na=False" we also get the "na" objects
```

```python
df[filt]
```

```python
df.loc[df["Country"].str.contains('s', case=False), "Player":"Blocks"]  # If True, case sensitive.

# using slicing gives error without loc
```
