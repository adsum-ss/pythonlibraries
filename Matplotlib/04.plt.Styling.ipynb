```python
# Basic Library
import numpy as np
import pandas as pd

# Libraries for Visualization
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

import warnings;
warnings.filterwarnings('ignore')
```

```python
from matplotlib import style
```

```python
x = np.arange(0,200,0.2)
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
style.use('ggplot')

x = np.arange(0,200,0.2)
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
style.use('ggplot')

x = np.arange(0,30,0.2)
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
style.use('fivethirtyeight')

x = np.arange(0,30,0.2)
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
style.use('dark_background')

x = np.arange(0,30,0.2)
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
style.use('grayscale')

x = np.arange(0,30,0.2)
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,30,0.2)
y = np.sin(x)

plt.title("Sine Function")
plt.suptitle("Students", y=1.05)

plt.xlabel("Weight of Students")
plt.ylabel("Height of Student")

plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,30,0.2)
y1 = np.sin(x)
y2 = np.cos(x)

plt.title("Sine Function")
plt.xlabel("Weight of Students")
plt.ylabel("Height of Student")

plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
x = np.arange(0,30,0.2)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1, label='Sine')
plt.plot(x,y2, label='Cosine')

plt.legend()
plt.show()
```
