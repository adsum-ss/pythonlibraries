## catplot

Figure-level interface for drawing categorical plots onto a FacetGrid.

This function provides access to several axes-level functions that show the relationship between a numerical and one or more categorical variables using one of several visual representations. The kind parameter selects the underlying axes-level function to use:

```python
# Basic Library
import numpy as np
import pandas as pd

# Libraries for Visualization
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

import scipy
import warnings;
warnings.filterwarnings('ignore')
```

```python
titanic = sns.load_dataset("titanic")
titanic
```

```python
sns.catplot(x="age", data=titanic);
```

```python
sns.catplot(y="age", data=titanic);
```

```python
sns.catplot(x="pclass", y="age", hue="sex", data=titanic);
```

```python
sns.catplot(x="pclass", y="age", hue="sex", data=titanic, col="survived");
```

```python
sns.catplot(x="pclass", y="age", hue="sex", data=titanic, col="survived", height=7, aspect=.8);
```

## kind = bar

```python
sns.catplot(x='pclass', y = "survived", data=titanic, kind="bar", estimator=np.count_nonzero);
```

```python
sns.catplot(x='pclass', y = "survived", data=titanic, kind="bar", estimator=np.count_nonzero, hue="sex");
```

```python
sns.catplot(x='pclass', y='survived', data=titanic, kind='bar', estimator=np.sum, col='sex');
```

```python
sns.catplot(x='pclass', y='survived', data=titanic, kind='bar', estimator=np.sum, hue='sex', col='alone', row='embarked');
```

```python
g = sns.catplot(x="who", y="survived", col="class", data=titanic, saturation=.5, kind="bar", ci=None, aspect=.6)

(g.set_axis_labels("", "Survival Rate")
  .set_xticklabels(["Men", "Women", "Children"])
  .set_titles("{col_name} {col_var}")
  .set(ylim=(0, 1))
  .despine(left=True)); 
```

## kind = count

```python
sns.catplot(x="alive", col="deck", col_wrap=4, data=titanic[titanic.deck.notnull()],
            kind="count", height=2.5, aspect=.8);
```

## kind = box

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='box');
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='box', hue='smoker');
```

```python
sns.set(style='darkgrid', font_scale=1.5)  

g = sns.catplot(x='day', y='total_bill', 
                data=tips,
                col='time',
                kind='box'
               )
g.set_titles(col_template="{col_name} Time");
```

```python
g = sns.catplot(x='day', y='total_bill', 
                data=tips,
                row='time',
                kind='box',
                aspect=1.5,
                height=4
               )
g.set_titles(row_template="{row_name} Time");
```

## kind = violin

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='violin', hue='smoker', facet_kws={'ylim':(0, 70)});
```

```python
sns.catplot(x="age", y="embark_town", hue="sex", row="class", data=titanic[titanic.embark_town.notnull()],
            orient="h", height=2, aspect=3, palette="Set3", kind="violin", dodge=True, cut=0, bw=.2);
```

## kind = swarm

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='swarm', hue='smoker');
```

## kind = strip

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='strip', hue='smoker');
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='strip', hue='smoker', aspect=1.8, s=5);
```

```python
sns.set_style('darkgrid')
sns.catplot(x='day', y='total_bill', data=tips, kind='strip', col='smoker', s=12, alpha=0.3, linewidth=2);
```

## kind = boxen

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='boxen', hue='smoker');
```

## kind = point

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='point', hue='smoker');
```
