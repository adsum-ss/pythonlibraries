## loc

<i> Access a group of rows and columns by label(s) or a boolean array.
   
<i> .loc[] is primarily label based, but may also be used with a boolean array.

```python
df.loc[0] # we accessed the first row by its label (series)
```

```python
df.loc[[0]] # we accessed the first row by its label (dataframe)
```

```python
df.loc[[0, 1]]
```

```python
df.loc[0, "embark_town"]
```

```python
df.loc[0]["embark_town"]
```

```python
df.loc[[0, 1], "embark_town"]
```

```python
df.loc[[0, 1, 2], ["embark_town"]]
```

```python
df.loc[[0, 1, 2], ["embark_town", "class"]] # for multiple columns
```

```python
df.loc[0:2, "embark_town"] # including stop point
```

```python
df.loc["class":"embark_town"]
```

```python
df.loc[0:2, "class":"embark_town"]  # including stop point

# when slicing, don't use square brackets
```

```python
df.loc[0:2, "class":"embark_town"][::2]
```

```python
df.loc[0:2][["who","embark_town"]]
```

## iloc

<i> Purely integer-location based indexing for selection by position.

<i> .iloc[] is primarily integer position based (from 0 to length-1 of the axis), but may also be used with a boolean array.

```python
df.iloc[0]  # first row
```

```python
df.iloc[[0, 1]]  # multiple rows
```

```python
df.iloc[[0, 1], 3]
```

```python
df.iloc[1:4, 1:4] # stop points not included
```

```python
df.iloc[0:4, [0,2,8,12]]  # don't use square brackets when slicing
```

## set_index

<i> Set the DataFrame index (row labels) using one or more existing columns or arrays (of the correct length). The index can replace the existing index or expand on it.

```python
df.index  # it is useful when we look for unique identifiers (but we don't have a unique identifier in this dataset)
```

```python
df.set_index("embark_town").head()  # it doesn't change the original dataframe
```

```python
df.head()
```

```python
df.set_index("embark_town", inplace=True)  # now it will modify the dataframe
df.head()
```

```python
df.loc["Cherbourg"] # if it was unique it would only return one result, however it can be used for grouping now.
```

```python
df.loc["Cherbourg", "class"].head()
```

```python
df.loc[0]  # we will get this error since the data frame has no longer index with that label
```

## reset_index

<i> Reset the index, or a level of it.

<i> Reset the index of the DataFrame, and use the default one instead. If the DataFrame has a MultiIndex, this method can remove one or more levels.

```python
df.reset_index(inplace=True)
df
```
