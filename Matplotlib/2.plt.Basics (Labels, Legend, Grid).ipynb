```python
# Basic Library
import numpy as np
import pandas as pd

# Libraries for Visualization
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

import warnings;
warnings.filterwarnings('ignore')
```

## Axes Labels and Chart Title

```python
age = [25, 27, 29, 31, 33, 35, 37, 39, 41, 43, 45]
salary = [38496, 42000, 46752, 49320, 53200, 56000, 62316, 64928, 67317, 68748, 73752]
salary2 = [45372, 48876, 53850, 57287, 63016, 65998, 70003, 70000, 71496, 75370, 83640]
```

```python
# functional method

plt.plot(age, salary)

plt.xlabel('X Axis Title Here', size=15, labelpad=15)
plt.ylabel('Y Axis Title Here', color='blue', loc='top')
plt.title('String Title Here', pad=15, size=15)

plt.show()
```

```python
# object-oriented method

fig, ax = plt.subplots()

ax.plot(age, salary)

ax.set_xlabel("X Axis Title Here")
ax.set_ylabel("Y Axis Title Here")
ax.set_title("String Title Here", size=15, pad=15);
```

```python
fig, ax = plt.subplots()

font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
        'ha': 'left',
        'va': 'center',
       }

ax.plot(age, salary)

ax.set_xlabel("X Axis Title Here", fontdict=font, labelpad=15)
ax.set_ylabel("Y Axis Title Here", fontdict=font, labelpad=15)
ax.set_title("String Title Here", fontdict=font, pad=15);
```

## Suptitle

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15,3))

fig.suptitle('Horizontally stacked subplots', size=20)

ax[0].plot(age, salary, 'orange')
ax[0].set_xlabel('age')
ax[0].set_title('First')

ax[1].plot(age, salary2, 'y')
ax[1].set_title('Second')
ax[1].set_xlabel('age')

plt.tight_layout();
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15,3))

ax[0].plot(age, salary, 'orange')
ax[0].set_xlabel('age')
ax[0].set_title('First')

ax[1].plot(age, salary2, 'y')
ax[1].set_title('Second')
ax[1].set_xlabel('age')

plt.suptitle('Horizontally stacked subplots', size=20, x=0.5, y=1.05)
plt.tight_layout();
```

## Legend

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
# Show legend

plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

plt.legend();
```
