## pairplot

Plot pairwise relationships in a dataset.

By default, this function will create a grid of Axes such that each numeric variable in data will by shared across the y-axes across a single row and the x-axes across a single column. The diagonal plots are treated differently: a univariate distribution plot is drawn to show the marginal distribution of the data in each column.

It is also possible to show a subset of variables or plot different variables on the rows and columns.

This is a high-level interface for PairGrid that is intended to make it easy to draw a few common styles. You should use PairGrid directly if you need more flexibility.

```python
# Basic Library
import numpy as np
import pandas as pd

# Libraries for Visualization
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

import scipy

import warnings;
warnings.filterwarnings('ignore')
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.pairplot(tips);
```

```python
tips.dtypes
```

```python
sns.pairplot(tips, hue='sex');
```

```python
sns.set_style('darkgrid')
sns.pairplot(tips, kind='kde');
```

```python
sns.pairplot(tips, hue='sex', palette='viridis', diag_kind='hist');
```

```python
sns.pairplot(tips, hue='sex', palette='viridis', corner=True, diag_kind='hist');
```

### Returns PairGrid

```python
sns.set_style('darkgrid')

g = sns.pairplot(tips)
```

```python
type(g)
```

```python
g = sns.pairplot(tips)
g.map_upper(sns.kdeplot, n_levels=6, color='xkcd:wine red');
```

```python
g = sns.pairplot(tips, diag_kind="kde")
g.map_lower(sns.kdeplot, levels=4, color=".2");
```

### Categorical and Specific Variables

```python
sns.pairplot(tips, hue='sex', vars=['total_bill', 'tip']);
```

```python
sns.pairplot(tips, hue='sex', 
             y_vars=['tip'],
             x_vars=['total_bill', 'size'],
             kind='reg', height=4);
```

### Styling

```python
sns.pairplot(tips, hue='sex', markers=["s", "D"], height=1.5);
```

```python
sns.pairplot(tips, diag_kws={'color': 'gray', 'alpha': 0.6});
```

```python
sns.pairplot(tips, diag_kind='kde', kind='reg', diag_kws={'color': 'black', 'shade':False, 'lw': 4})
plt.tight_layout();
```

```python
sns.pairplot(tips,  plot_kws={'color':'black'})
plt.tight_layout();
```

```python
sns.pairplot(tips, kind='reg', plot_kws={'line_kws': {'color':'black', 'lw': 5}})
plt.tight_layout();
```

```python
sns.pairplot(tips, kind='reg', plot_kws={'ci':None, 'color': 'xkcd:salmon', 'scatter_kws': {'color': 'gray', 's': 3}});
```

```python
sns.pairplot(
    tips,
    plot_kws=dict(marker="+", linewidth=1),
    diag_kws=dict(fill=False),
);
```
